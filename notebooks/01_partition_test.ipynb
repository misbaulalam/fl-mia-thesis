```python
import torchvision
import torchvision.transforms as transforms

transform = transforms.ToTensor()

train_dataset = torchvision.datasets.CIFAR10(
    root='../data',
    train=True,
    download=True,
    transform=transform
)
```

```python
import sys
sys.path.append('../src')
from partition import dirichlet_partition

train_labels = [label for _, label in train_dataset]

client_data = dirichlet_partition(train_labels, num_clients=5, alpha=0.5)

for client_id, indices in client_data.items():
    print(f"Client {client_id}: {len(indices)} samples")
```

```python
import numpy as np

def print_class_distribution(client_data, labels, num_clients, num_classes=10):
    labels = np.array(labels)
    for client_id in range(num_clients):
        client_labels = labels[client_data[client_id]]
        counts = [np.sum(client_labels == c) for c in range(num_classes)]
        print(f"Client {client_id}: {counts}")

print_class_distribution(client_data, train_labels, num_clients=5)
```

```python
# very skewed
client_data_low = dirichlet_partition(train_labels, num_clients=5, alpha=0.1)
print("=== alpha = 0.1 (very skewed) ===")
print_class_distribution(client_data_low, train_labels, num_clients=5)

print()

# nearly IID
client_data_high = dirichlet_partition(train_labels, num_clients=5, alpha=100)
print("=== alpha = 100 (nearly IID) ===")
print_class_distribution(client_data_high, train_labels, num_clients=5)
```
